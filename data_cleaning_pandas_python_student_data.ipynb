```python
df.loc[:, "name"] = df["name"].str.lower().str.title()
```

```python
age_mean = df["age"].mean().round(0)
```

```python
df.loc[:, "age"] = df["age"].fillna(age_mean)
```

```python
df
```

```python
df["gender"].unique()
```

```python
df.loc[:, "gender"] = df["gender"].str.lower().str.title()
```

```python
df
```

```python
df["grade"]
```

```python
df["grade"].unique()
```

```python
df.loc[:, "grade"] = df["grade"].replace({'10th':'10', '11th':'11'})
```

```python
df
```

```python
math_mean = df["math_score"].mean()
```

```python
df.loc[:, "math_score"] = df["math_score"].fillna(math_mean)
```

```python
df
```

```python
df["english_score"]
```

```python
df.loc[:, "english_score"] = pd.to_numeric(df["english_score"], errors='coerce')
```

```python
df
```

```python
df["enrolled_date"].unique()
```

```python
## yyyy-mm-dd
```

```python
dict_date={
    '10-06-2022':'2022-06-10',
    '06/12/2022':'2022-06-12', 
    '2022/06/11':'2022-06-11'
}
```

```python
df.loc[:, "enrolled_date"] = df["enrolled_date"].replace(dict_date)
```

```python
df["enrolled_date"] = pd.to_datetime(df["enrolled_date"])
```

```python
df["remarks"].unique()
```

```python
dict_remark = {
    'excellent': 'Excellent', 
    'GOOD': 'Good', 
    'needs improvement': 'Needs Improvement', 
    'average': 'Average', 
    'poor': 'Poor', 
    'good student': 'Good'
}
```

```python
df["remarks"].replace(dict_remark).unique()
```

```python
df.loc[:,"remarks"] = df["remarks"].replace(dict_remark)
```

```python
df
```

```python
# 1. Calculate Total Score
df['total_score'] = df['math_score'] + df['english_score']
```

```python
# 2. Gender-Based Performance Analysis
gender_performance = df.groupby('gender')[['math_score', 'english_score']].mean()
gender_performance
```

```python
# 3. Filter Students by Age (e.g., older than 15)
older_students = df[df['age'] > 15]
older_students
```

```python
# 4. Remarks Distribution Summary
remarks_summary = df['remarks'].value_counts()
remarks_summary
```

```python
# 5. Export Cleaned Data to CSV
df.to_csv('cleaned_student_data.csv', index=False)
```
