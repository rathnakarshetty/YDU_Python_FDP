```python
# Import Matplotlib
import matplotlib.pyplot as plt
```

```python
# Time (Months)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

# Sales values
sales = [200, 250, 300, 280, 350, 400]
```

```python
# Create line chart
plt.plot(months, sales)

# Show plot
plt.show()
```

```python
plt.plot(months, sales, color='blue', marker='o')

plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")

plt.show()
```
