```python
# Import required libraries
import pandas as pd          # For data handling
import matplotlib.pyplot as plt   # For visualization
```

```python
# Load the diabetes dataset from CSV file
data = pd.read_csv('diabetes.csv')

# Display first few rows to understand data
print(data.head())
```

# Check missing/invalid values

```python
# Check if there are missing values
print(data.isnull().sum())
```

# Create Boxplot

### Boxplot summarizes distribution in one compact form.

```python
# Separate groups
non_diabetic = data[data['Outcome'] == 0]['Glucose']
diabetic = data[data['Outcome'] == 1]['Glucose']
```

```python
plt.boxplot([non_diabetic, diabetic], tick_labels=['Non-Diabetic', 'Diabetic'])

plt.xlabel("Group")
plt.ylabel("Glucose Level")
plt.title("Glucose Distribution by Diabetes Outcome")

plt.show()
```
