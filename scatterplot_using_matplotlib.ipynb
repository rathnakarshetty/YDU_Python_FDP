```python
# Import Matplotlib library for plotting graphs
import matplotlib.pyplot as plt
```

```python
# Data for plotting
# X-axis values (Study hours of students)
study_hours = [1, 2, 3, 4, 5, 6, 7]

# Y-axis values (Marks scored by students)
marks = [40, 50, 55, 65, 70, 75, 85]
```

# Basic scatter plot

```python
# Create a basic scatter plot
plt.scatter(study_hours, marks)

# Display the graph
plt.show()
```

# Add X-axis label

```python
# Create scatter plot
plt.scatter(study_hours, marks)

# Add label to X-axis
plt.xlabel("Study Hours")

# Display the graph
plt.show()
```

# Add Y-axis label

```python
# Create scatter plot
plt.scatter(study_hours, marks)

# Add label to X-axis
plt.ylabel("Marks")

# Display the graph
plt.show()
```

```python
# Create scatter plot
plt.scatter(study_hours, marks)

# Add label to X-axis
plt.xlabel("Study Hours")

# Add label to Y-axis
plt.ylabel("Marks")

# Display the graph
plt.show()
```

# Add Title

```python
# Create scatter plot
plt.scatter(study_hours, marks)

# Add labels to axes
plt.xlabel("Study Hours")
plt.ylabel("Marks")

# Add title to the graph
plt.title("Study Hours vs Marks")

# Display the graph
plt.show()
```

# Change color

```python
# Create scatter plot with custom color
plt.scatter(study_hours, marks, color='green')
# try other colors (blue, black, yello)

# Add labels and title
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

# Display the graph
plt.show()
```

# Change marker style

```python
# Create scatter plot with custom color and marker style
plt.scatter(study_hours, marks, color='blue', marker='x')
# try other colors (+, *, s, d, p, h)

# Add labels and title
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

# Display the graph
plt.show()
```

# Add grid

```python
# Create scatter plot
plt.scatter(study_hours, marks)

# Add labels and title
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

# Add grid lines for better readability
plt.grid()

# Display the graph
plt.show()
```
